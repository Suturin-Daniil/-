```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv('Cl.csv')
alpha = data1.iloc[:,0].to_numpy()
Cl = data1.iloc[:,1].to_numpy()

data2 = pd.read_csv('Cd.csv')
Cd = data2.iloc[:,1].to_numpy()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,4))
ax = ax.ravel()
ax[0].scatter(alpha, Cl, s=3)
ax[0].grid('Major')
ax[0].set_xlim(0, 13)
ax[0].set_ylim(0, 1.2)
ax[0].set_xlabel(r'$\alpha^{0}$')
ax[0].set_ylabel(r'$C_{l}$')
ax[0].set_title(r'Зависимость $C_{l}$ от угла атаки')
ax[0].plot([3, 3], [0, 1.2], '--', color='orange')
ax[1].scatter(alpha, Cd, s=3)
ax[1].set_xlim(0, 12)
ax[1].set_ylim(11, 22)
ax[1].grid('Major')
ax[1].set_xlabel(r'$\alpha^{0}$')
ax[1].set_ylabel(r'$\frac{C_{l}}{C_{d}}$')
ax[1].set_title(r'Зависимость $\frac{C_{l}}{C_{d}}$ от угла атаки')
ax[1].plot([3, 3], [11, 22], '--', color='orange')
plt.savefig('XFLRStudy.png', dpi=300, pad_inches = 0, bbox_inches = 0)
```
