```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([11.75, 10.5, 9, 8.8, 8.6, 6.6, 6.1, 5.6, 5])
d = np.array([-2, 0, 1, 2, 3, 5, 7, 9, 11])

p = np.polyfit(d,a, 1)

fig, ax = plt.subplots(1,1, figsize=(12,4))
ax.scatter(d, a, c='orange', s=10)
ax.plot(np.linspace(-5,20), np.polyval(p, np.linspace(-5,20)))
ax.set_xlim(-5, 20)
ax.set_ylim(0,14)
ax.grid('Major')
ax.set_xlabel(r'$\delta_{Ð²}, ^{o}$')
ax.set_ylabel(r'$\alpha, ^{o}$')
plt.savefig('BalanceCurve.png', dpi=300, bbox_inches='tight', pad_inches=0)
```
